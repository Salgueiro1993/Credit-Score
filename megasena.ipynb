```python
import random

def sortear_numeros(qtd_numeros):
    numeros_sorteados = []
    while len(numeros_sorteados) < qtd_numeros:
        numero = random.randint(1, 60)
        if numero not in numeros_sorteados:
            numeros_sorteados.append(numero)
    return numeros_sorteados

def main():
    qtd_numeros = 6  # Quantidade de números a serem sorteados
    numeros_sorteados = sortear_numeros(qtd_numeros)
    print("Números sorteados:")
    print(numeros_sorteados)

if __name__ == "__main__":
    main()
```

```python
import random

def generate_lotofacil_game():
    return random.sample(range(1, 26), 15)

num_games = 6

print("Seus 6 jogos da Lotofácil:")

for game in range(num_games):
    numbers = generate_lotofacil_game()
    print(f"Jogo {game+1}: {numbers}")
```
